```python
import pandas as pd

# Nome do arquivo para verificar
path = "perfil_eleitorado_2020.csv"

# Nome do arquivo para ser gerado
resultPath = "Resultado_filtro.csv"

# Nome da coluna que deseja ver a condição
targetColumnName = "CD_GRAU_ESCOLARIDADE" 

# Resultado que deseja conseguir em referencia
target = 4


df_sample = pd.read_csv(path, nrows=10)
df_sample_size = df_sample.memory_usage(index=True).sum()

my_chunk = (1000000000 / df_sample_size)/10 
my_chunk = int(my_chunk//1)
# Cria o csv
iter_csv = pd.read_csv(
    path, 
    iterator=True, 
    chunksize=my_chunk,
    delimiter=";")
# Faz o filtro
df_result = pd.concat(
    [chunk[chunk[targetColumnName]==target] 
    for chunk in iter_csv])

df_result.to_csv(resultPath,sep=";")
print("Feito!")
```
